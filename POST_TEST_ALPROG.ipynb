```python
# 1. Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

GRAFIK 1

```python
# 1. Membaca Dataset
df = pd.read_csv('Kelas G_Fixed Call Report.csv')

# 2. Pengolahan Data (Mencari Standar Deviasi per Tahun)
df_grafik1 = df.groupby('Year')['Total Number of Calls'].std().reset_index()

# 3. Visualisasi Data (Membuat Bar Chart)
fig, ax = plt.subplots(figsize=(8, 5))

# Membuat Bar Chart menggunakan Seaborn
sns.barplot(data=df_grafik1, x='Year', y='Total Number of Calls', ax=ax, palette='viridis')

# Menambahkan judul, label, dan grid agar sesuai standar profesional tugas lu
ax.set_title('Standar Deviasi Total Panggilan Masuk Bulanan per Tahun', fontsize=12, fontweight='bold')
ax.set_xlabel('Tahun', fontsize=10)
ax.set_ylabel('Standar Deviasi', fontsize=10)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

GRAFIK 2

```python
# 1. Filter data khusus untuk tahun 2024
df_2024 = df[df['Year'] == 2024].copy()

# 2. Mengurutkan bulan agar grafiknya rapi dari Januari - Desember
urutan_bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
df_2024['Month'] = pd.Categorical(df_2024['Month'], categories=urutan_bulan, ordered=True)
df_2024 = df_2024.sort_values('Month')

# 3. Membuat wadah grafik (figure dan axes)
fig, ax = plt.subplots(figsize=(10, 5))

# 4. Membuat Multi-Line Chart
# Garis 1: Konsultasi Dokter
sns.lineplot(data=df_2024, x='Month', y='Total Number of Doctors Consultancy', 
             marker='o', label='Konsultasi Dokter', ax=ax, color='b')

# Garis 2: Informasi Ambulans
sns.lineplot(data=df_2024, x='Month', y='Number of Total Ambulance Information', 
             marker='s', label='Informasi Ambulans', ax=ax, color='r')

# 5. Merapikan tampilan grafik (Judul, Label, dan Grid)
ax.set_title('Tren Bulanan Layanan Kesehatan Masyarakat (Tahun 2024)', fontsize=12, fontweight='bold')
ax.set_xlabel('Bulan', fontsize=10)
ax.set_ylabel('Jumlah Panggilan', fontsize=10)
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend(title='Jenis Layanan') # Menampilkan legenda garis

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

GRAFIK 3

```python
# 1. Membuat wadah grafik (figure dan axes)
fig, ax = plt.subplots(figsize=(8, 5))

# 2. Membuat Scatter Plot menggunakan Seaborn
sns.scatterplot(
    data=df, 
    x='Number of Total Ambulance Information', 
    y='Total Number of Calls', 
    ax=ax, 
    color='purple', 
    alpha=0.6,
    edgecolor=None
)

# 3. Merapikan tampilan grafik (Judul, Label, dan Grid)
ax.set_title('Korelasi Informasi Ambulans vs Total Panggilan Masuk', fontsize=12, fontweight='bold')
ax.set_xlabel('Jumlah Panggilan Informasi Ambulans', fontsize=10)
ax.set_ylabel('Total Panggilan Masuk Keseluruhan', fontsize=10)
ax.grid(True, linestyle='--', alpha=0.6)

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

GRAFIK 4

```python
# 1. Membuat wadah grafik (figure dan axes)
fig, ax = plt.subplots(figsize=(8, 5))

# 2. Membuat Histogram menggunakan Seaborn
sns.histplot(
    data=df, 
    x='Number of Calls To Know About The Service', 
    bins=15, # Membagi data ke dalam 15 rentang agar sebarannya terlihat jelas
    ax=ax, 
    color='mediumseagreen',
    edgecolor='black'
)

# 3. Merapikan tampilan grafik (Judul, Label, dan Grid)
ax.set_title('Distribusi Frekuensi Panggilan Pencarian Info Layanan', fontsize=12, fontweight='bold')
ax.set_xlabel('Jumlah Panggilan', fontsize=10)
ax.set_ylabel('Frekuensi (Bulan)', fontsize=10)
ax.grid(axis='y', linestyle='--', alpha=0.6)

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

GRAFIK 5

```python
# 1. Membuat wadah grid 2 baris x 2 kolom
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Menambahkan judul utama untuk keseluruhan Dashboard
fig.suptitle('Dashboard Analisis Data Panggilan Layanan Kesehatan', fontsize=16, fontweight='bold', y=1.02)

# ==========================================
# GRAFIK 1 (Kiri Atas): axes[0, 0] - Agregasi
# ==========================================
sns.barplot(data=df_grafik1, x='Year', y='Total Number of Calls', ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Standar Deviasi Total Panggilan per Tahun', fontsize=11, fontweight='bold')
axes[0, 0].set_xlabel('Tahun', fontsize=9)
axes[0, 0].set_ylabel('Standar Deviasi', fontsize=9)
axes[0, 0].grid(axis='y', linestyle='--', alpha=0.7)

# ==========================================
# GRAFIK 2 (Kanan Atas): axes[0, 1] - Tren
# ==========================================
sns.lineplot(data=df_2024, x='Month', y='Total Number of Doctors Consultancy', marker='o', label='Konsultasi Dokter', ax=axes[0, 1], color='b')
sns.lineplot(data=df_2024, x='Month', y='Number of Total Ambulance Information', marker='s', label='Informasi Ambulans', ax=axes[0, 1], color='r')
axes[0, 1].set_title('Tren Bulanan Layanan Kesehatan (2024)', fontsize=11, fontweight='bold')
axes[0, 1].set_xlabel('Bulan', fontsize=9)
axes[0, 1].set_ylabel('Jumlah Panggilan', fontsize=9)
axes[0, 1].grid(True, linestyle='--', alpha=0.6)
axes[0, 1].legend()

# ==========================================
# GRAFIK 3 (Kiri Bawah): axes[1, 0] - Korelasi
# ==========================================
sns.scatterplot(data=df, x='Number of Total Ambulance Information', y='Total Number of Calls', ax=axes[1, 0], color='purple', alpha=0.6)
axes[1, 0].set_title('Korelasi Info Ambulans vs Total Panggilan', fontsize=11, fontweight='bold')
axes[1, 0].set_xlabel('Jumlah Panggilan Info Ambulans', fontsize=9)
axes[1, 0].set_ylabel('Total Panggilan Masuk', fontsize=9)
axes[1, 0].grid(True, linestyle='--', alpha=0.6)

# ==========================================
# GRAFIK 4 (Kanan Bawah): axes[1, 1] - Distribusi
# ==========================================
sns.histplot(data=df, x='Number of Calls To Know About The Service', bins=15, ax=axes[1, 1], color='mediumseagreen', edgecolor='black')
axes[1, 1].set_title('Distribusi Panggilan Pencarian Info Layanan', fontsize=11, fontweight='bold')
axes[1, 1].set_xlabel('Jumlah Panggilan', fontsize=9)
axes[1, 1].set_ylabel('Frekuensi (Bulan)', fontsize=9)
axes[1, 1].grid(axis='y', linestyle='--', alpha=0.6)

# ==========================================
# Merapikan dan Menampilkan Output
# ==========================================
plt.tight_layout() # Mencegah teks antar grafik saling tumpang tindih
plt.show()
```
